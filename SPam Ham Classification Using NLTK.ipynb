```python
import pandas as pd
messages = pd.read_csv("sms_spam.csv",names = ["label","message"])
```

```python
messages = messages.drop(0)
```

```python
messages = messages.reset_index(drop=True)
```

```python
messages.head()
```

```python
# Data Cleaning and preprocessing
import re 
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmitizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]',' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmitizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len(corpus)
```

```python
#  Output Features
y = pd.get_dummies(messages['label'])
y = y.astype(int)
y = y.iloc[:,0].values
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size = 0.2,random_state=42)
```

```python
# Create Bag of Words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 25000,ngram_range=(1,2))
```

```python
# Independend feature
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size = 0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train = tv.fit_transform(X_train).toarray()
X_test = tv.transform(X_test).toarray()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```
